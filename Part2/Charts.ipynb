```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data from CSV file
df = pd.read_csv("monthly_cohort_analysis.csv")
```

```python
# Ensure the cohort_month column is used as the index
df.set_index("cohort_month", inplace=True)
```

```python
# Calculate retention percentages
df_percentage = df.divide(df.iloc[:, 0], axis=0) * 100
```

```python
# 1. Retention Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_percentage, annot=True, fmt=".1f", cmap="Blues")
plt.title("Retention Heatmap")
plt.xlabel("Months After Signup")
plt.ylabel("Cohort Month")
plt.show()
```

```python
# 2. Cohort Size Bar Plot
plt.figure(figsize=(8, 5))
df.iloc[:, 0].plot(kind="bar", color="skyblue")
plt.title("Cohort Sizes")
plt.xlabel("Cohort Month")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.show()
```

```python

# 3. Retention Line Plot
plt.figure(figsize=(10, 6))
for cohort in df_percentage.index:
    plt.plot(df_percentage.columns, df_percentage.loc[cohort], label=cohort)

plt.title("Retention Over Time by Cohort")
plt.xlabel("Months After Signup")
plt.ylabel("Retention Percentage")
plt.legend(title="Cohort Month")
plt.grid(True)
plt.show()
```

```python
# 4. Cohort Growth Funnel
plt.figure(figsize=(8, 5))
for cohort in df.index:
    plt.plot(df.loc[cohort], label=f"{cohort}")

plt.title("Cohort Growth Funnel")
plt.xlabel("Months After Signup")
plt.ylabel("Number of Users")
plt.legend(title="Cohort Month")
plt.grid(True)
plt.show()
```
